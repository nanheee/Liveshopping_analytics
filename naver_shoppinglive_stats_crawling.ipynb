### 네이버 쇼핑라이브 통계 데이터 크롤링 

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import numpy as np 
import pandas as pd
import time
import random

id_num = []
live_title = []
live_datetime = []
live_status = []
payment_price = []
product_view = []
banner_view = []
payment_count = []
view_total = []
view_login = []
view_logout = []
view_max = []
alarm_total = []
alarm_new = []
man = []
man_purchase = []
woman = []
woman_purchase = []
user_chat = []
user_good = []
user_heart = []
created_date = []

created_date_1 = []
created_date_2 = []
id_num_1 = []
id_num_2 = []
product_name_1 = []
product_name_2 = []
pd_payment_price = []
pd_payment_count = []

def crawling():
    time.sleep(2)
    list = '1','4'
    for i in list:
        driver.find_element_by_xpath('//*[@id="content"]/div[2]/div[3]/div[2]/div['+i+']/div[1]/button').click()
        time.sleep(1)
        html = driver.page_source
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        time.sleep(1)
        name = soup.find('div',{'class':'GridGroup_touchscrollable_1JuCv'}).find_all('strong',{'PaymentPopup_name_3dIKi'})
        price = soup.find('div',{'class':'GridGroup_touchscrollable_1JuCv'}).find_all('div',{'GridCell_wrap_1dFX2 PaymentPopup_cell_count_2FTf6'})
        view = soup.find('div',{'class':'GridGroup_touchscrollable_1JuCv'}).find_all('div',{'GridCell_wrap_1dFX2 PaymentPopup_cell_count_2FTf6'})
        count = soup.find('div',{'class':'GridGroup_touchscrollable_1JuCv'}).find_all('div',{'GridCell_wrap_1dFX2 PaymentPopup_cell_count_2FTf6'})
        if i == '1':
            for j in range(0,len(name)):
                id_num_1.append(x)
                product_name_1.append(name[j].get_text())
                pd_payment_price.append(price[j].get_text())
                created_date_1.append(time.strftime("%Y-%m-%d"))
        elif i == '4':
            for j in range(0,len(name)):
                id_num_2.append(x)
                product_name_2.append(name[j].get_text())
                pd_payment_count.append(view[j].get_text())
                created_date_2.append(time.strftime("%Y-%m-%d"))
        driver.find_element_by_xpath('//*[@id="content"]/div[2]/div[3]/div[2]/div[3]/button').click()
    live_title.append(soup.find('h2',{'class':'StatisticsWrapper_title_KxBD9'}).get_text())
    live_datetime.append(soup.find('time',{'class':'StatisticsWrapper_time_2u-vj'}).get_text())
    live_status.append(soup.find('span',{'class':'StatisticsWrapper_text_bDU_f'}).get_text())
    payment_stat = soup.find('div',{'class':'PaymentStat_box_1vpBM'}).find_all('div',{'class':'PaymentStat_number_2CZDU'})
    payment_price.append(payment_stat[0].get_text())
    product_view.append(payment_stat[1].get_text())
    banner_view.append(payment_stat[2].get_text())
    payment_count.append(payment_stat[3].get_text())
    view_stat = soup.find('div',{'StatList_wrap_wOv3- StatList_type_vertical_3w-Nu ViewStat_list_3e3Lt'}).find_all('span',{'class':'EllipsisText_ellipsis_2y20w'})
    view_total.append(view_stat[0].get_text())
    view_login.append(view_stat[1].get_text())
    view_logout.append(view_stat[2].get_text())
    view_max.append(view_stat[3].get_text())
    alarm_total.append(view_stat[4].get_text())
    alarm_new.append(view_stat[5].get_text())
    purchase_stat = soup.find('div',{'UserWatchAndPurchaseStat_legend_2tBHL'}).find_all('span',{'class':'blind'})
    user_action  = soup.find('div',{'StatBox_wrap_sSgf4 UserActionStat_wrap_2egyA'}).find_all('div',{'class':'StatList_number_3xNlR'})
    user_chat.append(user_action[0].get_text())
    user_good.append(user_action[1].get_text())
    user_heart.append(user_action[2].get_text())  
    id_num.append(x)
    created_date.append(time.strftime("%Y-%m-%d"))

driver = webdriver.Chrome('chromedriver.exe')
driver.implicitly_wait(3)
url = 'https://tool.shoppinglive.naver.com/'
driver.get(url)
#login
```
